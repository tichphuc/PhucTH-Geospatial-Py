```python
import geospatialtools as gst
```

```python
import geopandas as gpd

df_meteo_stations = gpd.read_file(r"F:\202507_Flood_TTHue\data\meteo\stations_label.shp")
df_meteo_stations
```

```python
meteo_features = ['precipitation', 'et0_fao_evapotranspiration', 'soil_moisture_0_to_7cm', 'soil_moisture_7_to_28cm']
se_time = ["1999-10-25", "1999-11-12"]
timezone = "Asia/Bangkok"


output_folder = r'F:\202507_Flood_TTHue\data\meteo\raw'

for id, coor in enumerate(df_meteo_stations.geometry.values[3:]):
    print(f"ID: {id:02}, Tọa độ: {coor}")

    output_path = output_folder + '\\ '+f"{id:02}.csv"

    gst.getMeteoData(
        [coor.y, coor.x],
        se_time,
        meteo_features,
        timezone=timezone,
        output_path=output_path
    )
```
